```python
import requests
from bs4 import BeautifulSoup
import mysql.connector


def extract_title(website_url):
    try:
        # Send an HTTP request to the website
        response = requests.get(website_url)
        response.raise_for_status()  # Check for any errors during the request
        
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find the <title> tag
        title_tag = soup.find('title')
        
        if title_tag:
            return title_tag.text.strip()
        else:
            return None
    except requests.exceptions.RequestException as e:
        print(f"Error fetching website content: {e}")
        return None

# Function to store the website name and title in MySQL database
def store_in_mysql(website_url, title):
    try:
        # Connect to MySQL database
        conn = mysql.connector.connect(
            host='localhost',  # Update with your MySQL host if needed
            user='root',       # Update with your MySQL username
            password='1234',  # Update with your MySQL password
            database='web_data'  # Update with your database name
        )
        
        cursor = conn.cursor()
        
        # Create table if it doesn't exist
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS websites (
                id INT AUTO_INCREMENT PRIMARY KEY,
                website_name VARCHAR(255) NOT NULL,
                title VARCHAR(255) NOT NULL
            )
        ''')
        
        # Insert website and title data into the table
        cursor.execute('''
            INSERT INTO websites (website_name, title)
            VALUES (%s, %s)
        ''', (website_url, title))
        
        # Commit changes and close the connection
        conn.commit()
        print(f"Successfully saved {website_url} and its title.")
    except mysql.connector.Error as err:
        print(f"Error connecting to MySQL: {err}")
    finally:
        if conn.is_connected():
            cursor.close()
            conn.close()

# Main function to execute the program
def main():
    website_url = input("Enter the website URL: ").strip()
    
    # Get the title of the website
    title = extract_title(website_url)
    
    if title:
        print(f"Title of the website '{website_url}': {title}")
        
        # Store website URL and title in MySQL
        store_in_mysql(website_url, title)
    else:
        print(f"Could not extract title from {website_url}")

if __name__ == "__main__":
    main()
```
